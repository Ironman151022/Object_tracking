```python
import numpy as np
m = 4
dt = 0.1
std_a = 2
```

```python
A = np.eye(2*m,2*m) # State Transistion Matrix - A (2*m,2*m) 
for i in range(m):
    A[i][m+i] = dt
```

```python
B = np.zeros((2*m,m)) # Control Matrix - B (2*m,m)
for i in range(m):
    B[i][i] = (dt**2)/2
    B[i+m][i] = (dt)
```

```python
H = np.zeros((m,2*m)) # Transformation Matrix - H (m,2*m)
for i in range(m):
    H[i][i] = 1
```

```python
Q = np.zeros((2*m,2*m)) # Process Covariance Matrix - q (2*m,2*m)
for i in range(m):
    Q[i][i] = (dt**4)/4
    Q[i][i+m] = (dt**3)/2
    Q[i+m][i] = (dt**3)/2
    Q[i+m][i+m] = dt**2
Q = Q*(std_a**2)

R = np.zeros((m,m)) # Measurement Noise Covariance - R (m,m)
for i in range(m):
    R[i][i] = (dt**4)/4
R = R*(std_a**2)

X = np.zeros((2*m,1))
U = np.ones((m,1))
P = np.eye(A.shape[1])
```

```python
P.shape
```

```python
X = np.dot(A,X) +np.dot(B,U)
P = np.dot(np.dot(A,P),A.T) + Q
```
